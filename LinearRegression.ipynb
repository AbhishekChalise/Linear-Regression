```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
california = fetch_california_housing()
data = pd.DataFrame(california.data , columns = california.feature_names)
data['DESCR'] = california.target
```

```python
data.head()
```

```python
# Since there is no null in the data there is no problem we can proceed further.
data.isnull().sum()
```

```python
# NOw the data is split into train test and validation data
from sklearn.model_selection import train_test_split
X = data.drop('DESCR',axis = 1)
y = data['DESCR']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 , random_state = 40)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
len(predictions)
```

```python
len(y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MAE  =  mean_absolute_error( y_test , predictions )
MSE = mean_squared_error( y_test , predictions )
R2S = r2_score( y_test , predictions )
```

```python
print("Mean Absolute Error: ",MAE)
print("Mean Square Error: ",MSE)
print("R2_score: ",R2S)
```

### ASSUMPTIONS FOR LINEAR REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
"DATA"
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
data = pd.read_csv(url, names=column_names, delim_whitespace=True, na_values='?')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

## 1.Linearity

```python
sns.pairplot(data, x_vars=['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'], y_vars='mpg', height=4, aspect=0.8, kind='reg')
plt.show()
```

## 2. Independence of Errors

```python
from statsmodels.stats.stattools import durbin_watson
import statsmodels.api as sm

# Define the predictors and the target variables
X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
y = data['mpg']
# Add a constant to the predictors
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()
residuals = model.resid

# Durbin-Watson test
dw_stat = durbin_watson(residuals)
dw_stat
```

- Value close to 0 : Strong Positive AutoCorrelation , Don't follow independence of Errors- 
Value close to 2 : No AutoCorrelation , Follow Independence of Error
- 
Value close to  4: Strong Negative AutoCorrelation , Don't follow independence oE errors

### 3.Homoscedasticity

```python
# Residuals plot
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted values')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.show()
```

- Homodasecity : If points are scattered randomly and dont follow a pattern

# 4. Normality of Errors

```python
plt.hist(model.resid, bins=30)
plt.title('Histogram of residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()
```

# 5. No Multicollinearity

```python
X.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

vif_data
```

- Value above 10 : High Multicollinearity
